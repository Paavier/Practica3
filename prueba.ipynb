```python
import pandas as pd
import numpy as np
from sklearn.metrics import roc_auc_score
```

```python
model_A_df = pd.read_csv('data/Model_A.csv')
model_B_df = pd.read_csv('data/Model_B.csv')
model_C_df = pd.read_csv('data/Model_C.csv')
model_D_df = pd.read_csv('data/Model_D.csv')
```

```python
def TN(dataframe, threadshold):
    return ((dataframe['gt'] == 0) & (dataframe['pred'] < threadshold)).sum()
```

```python
def TP (dataframe, threadshold):
   return ((dataframe['gt'] == 1) & (dataframe['pred'] >= threadshold)).sum()
```

```python
def FN (dataframe, threadshold):
    return ((dataframe['gt'] == 1) & (dataframe['pred'] < threadshold)).sum()
```

```python
def FP (dataframe, threadshold):
    return ((dataframe['gt'] == 0) & (dataframe['pred'] >= threadshold)).sum()
```

```python
def Accuracy (dataframe, threadshold):
    tp = TP(dataframe, threadshold)
    tn = TN(dataframe, threadshold)
    fp = FP(dataframe, threadshold)
    fn = FN(dataframe, threadshold)
    
    return (tp + tn)/(tp + tn + fp + fn)
```

```python
def Precision(dataframe, threadshold):
    tp = TP(dataframe, threadshold)
    fp = FP(dataframe, threadshold)
    return tp/(tp + fp)
```

```python
def Recall(dataframe, threadshold):
    tp = TP(dataframe, threadshold)
    fn = FN(dataframe, threadshold)
    return tp/(tp + fn)
```

```python
def TNR(dataframe, threadshold):
    tn = TN(dataframe, threadshold)
    fp = FP(dataframe, threadshold)
    return tn/(tn + fp)
```

```python
def FPR(dataframe, threadshold):
    fp = FP(dataframe, threadshold)
    tn = TN(dataframe, threadshold)
    return fp/(fp + tn)
```

```python
def F1_score(dataframe, threadshold):
    precision = Precision(dataframe, threadshold)
    recall = Recall(dataframe, threadshold)
    return 2*(precision * recall)/(precision + recall)
```

```python
def Auc (dataframe):
    return roc_auc_score(dataframe['gt'], dataframe['pred'])
```

```python
#Metricas de modelo los modelos
threadshold = 0.5
datasets = {'Dataset_A' : model_A_df,
            'Dataset_B' : model_B_df,
            'Dataset_C' : model_C_df,
            'Dataset_D' : model_D_df}

results = []

for name, df in datasets.items():
    tn = TN(df, threadshold)
    tp = TP(df, threadshold)
    fn = FN(df, threadshold)
    fp = FP(df, threadshold)
    accuracy = Accuracy(df, threadshold)
    recall = Recall(df, threadshold)
    precision = Precision(df, threadshold)
    tnr = TNR(df, threadshold)
    f1_score = F1_score(df, threadshold)
    auc = Auc(df)
    
    results.append({
        'Dataset' : name,
        'TN' : tn,
        'TP' : tp,
        'FN' : fn,
        'FP' : fp,
        'ACCURACY': accuracy,
        'RECALL' : recall,
        'PRECISION' : precision,
        'TNR' : tnr,
        'F1-SCORE': f1_score,
        'AUC' : auc
    })

df_metricas = pd.DataFrame(results)
print(df_metricas)
```

```python
import pandas as pd
import numpy as np
```

```python
model_0_df = pd.read_csv('data/Reg_0.csv')
model_1_df = pd.read_csv('data/Reg_1.csv')
model_2_df = pd.read_csv('data/Reg_2.csv')
model_3_df = pd.read_csv('data/Reg_3.csv')
model_4_df = pd.read_csv('data/Reg_4.csv')
```

```python
def calc_mae(df):
    return (df['gt'] - df['pred']).abs().mean()
```

```python
def calc_mse(df):
    return ((df['gt'] - df['pred']) ** 2).mean()
```

```python
def calc_medAE(df):
    return (df['gt'] - df['pred']).abs().median()
```

```python
def calc_rmse(df):
    return np.sqrt(calc_mse(df))
```

```python
#Metricas de modelo los modelos
datasets = {'Dataset_0' : model_0_df,
            'Dataset_1' : model_1_df,
            'Dataset_2' : model_2_df,
            'Dataset_3' : model_3_df,
            'Dataset_4' : model_4_df}

results = []

for name, df in datasets.items():
    mae = calc_mae(df)
    mse = calc_mse(df)
    medAE = calc_medAE(df)
    rmse = calc_rmse(df)
    
    results.append({
        'Dataset' : name,
        'MAE' : mae,
        'MSE' : mse,
        'medAE' : medAE,
        'RMSE' : rmse,
    })

df_metricas = pd.DataFrame(results)
print(df_metricas)
```
